## import plotting library

```python
import matplotlib.pyplot as plt
```

## create basic Rectangle class

```python
class Rect(object):
    def __init__(self, cx, cy, width, height):
        self.cx = cx
        self.cy = cy
        self.width = width
        self.height = height
        w2 = self.width/2.0
        h2 = self.height/2.0
        # create a list of x and a list of y coordinates for this rect
        # last point same as first
        self.x = [self.cx-w2, self.cx+w2, self.cx+w2, self.cx-w2, self.cx-w2]
        self.y = [self.cy-h2, self.cy-h2, self.cy+h2, self.cy+h2, self.cy-h2]
    
    def display(self):
        plt.axis('equal')
        # uncomment this line for xkcd comic style line wiggles
        #plt.xkcd()
        plt.plot(self.x, self.y, c='k', linewidth=1)
```

## create and display one rectangle

```python
r = Rect(1,2,3,4)
r.display()
```

## create a list of 25 rectangles in a 5 x 5 grid, all of equal size

```python
rects = []
for x in range(5):
    for y in range(5):
        r = Rect(x*10, y*10, 8,8)
        rects.append(r)
```

## ...and display them

```python
for r in rects:
    r.display()
```

## import the random module

```python
import random
```

## randomly modify the size $s$ of the rectangles

```python
rects = []
for x in range(5):
    for y in range(5):
        s = 3+random.random()*6
        r = Rect(x*10, y*10, s,s)
        rects.append(r)

for r in rects:
    r.display()
```

## create 5 different rectangles at each grid position

```python
# make bigger and square plot
plt.figure(figsize=(12,12))
rects = []
for x in range(5):
    for y in range(5):
        for i in range(5):
            s = 3+random.random()*6
            r = Rect(x*10, y*10, s,s)
            rects.append(r)

for r in rects:
    r.display()
```

## randomly offset the rectangles' centers by $ox$ and $oy$

```python
plt.figure(figsize=(12,12))
plt.axis('off')
rects = []
for x in range(5):
    for y in range(5):
        for i in range(5):
            s = 3+random.random()*6
            ox = random.random()*3 - 1.5
            oy = random.random()*3 - 1.5
            r = Rect(x*10 + ox, y*10 + oy, s,s)
            rects.append(r)

for r in rects:
    r.display()
```
